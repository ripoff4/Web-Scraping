```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('quotestoscrape.csv')
df.head()
```

```python
top_authors = df['author'].value_counts()
```

```python
from collections import Counter
all_tags = df['tags'].dropna().str.split(',').sum()
tag_counts = Counter(all_tags)
top_tags = pd.Series(tag_counts).sort_values(ascending=False)
```

```python
print("Top Tags")
print(top_tags.head(10))
```

```python
print("Top Authors")
print(top_authors.head(10))
```
